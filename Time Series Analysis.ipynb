```python
sns.histplot(data= df , y ='#Passengers' , bins = 40)
```

```python
df_mul_decompose = seasonal_decompose(df , model = "multiplicative")
df_mul_decompose.plot()
plt.show()
```

```python
#to convert multiplicative model into additive model we to take log 
```

```python
df_log = df.copy()
df_log['#Passengers'] = np.log(df)
```

```python
#df_log.#Passengers
```

```python
df_log.plot()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.title('Original Time Series')
plt.plot(df1)

plt.subplot(2,1,2)
plt.title('Log Transformed Series')
plt.plot(df_log)
plt.tight_layout()
```
